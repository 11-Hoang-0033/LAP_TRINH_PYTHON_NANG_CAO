### ý 1c

```python
companies = pd.read_csv('D:/Advanced_Programming_Exercises/CHUONG_4/companies.csv')
companies.dtypes
companies.info
```

### ý 2

```python
stocks1_na = pd.isnull(stocks1)
print(stocks1_na, "\nlà kiểm tra có dữ liệu null trong stocks1 không")
```

### ý 3

```python
stocks = pd.concat([stocks1,stocks2])
stocks.tail(15)
```

### ý 4

```python
stocks_companies = pd.concat([stocks,companies])
stocks_companies.head(5)
```

### ý 5

```python
gia_tb = stocks_companies.groupby('name').agg({
    'open':'mean',
    'high':'mean',
    'low':'mean',
    'close':'mean'
})
print(gia_tb, "\nlà giá trung bình các công ty")

volume_tb =stocks_companies.groupby('name').agg({
    'volume':'mean'
})
print(volume_tb, "\nlà volume trung bình các công ty")
```

### ý 6

```python
stocks_companies.groupby('name').agg({
    'close':'mean',
    'close':'max',
    'close':'min'
})
```

### ý 7

```python
parsed_time = pd.to_datetime(stocks_companies.date)
type(parsed_time)
stocks_companies = pd.concat([stocks_companies,parsed_time],ignore_index=True)
stocks_companies.head(5)
```
